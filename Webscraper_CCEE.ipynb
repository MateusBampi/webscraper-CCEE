```python
# -*- coding: utf-8 -*-

"""
Created on Tue May 17 15:02:41 2022

@author: U350527 & U342925

Para que este código funcione, é necessário capturar as imagens que constam na mesma pasta desse arquivo

Antes de iniciar: 
   1- Abrir o site https://www.ccee.org.br/pt/dados-e-analises/seguranca-de-mercado/indicadores-de-seguranca
   2- Colocar o zoom em 50% de forma que toda a tela do Tableau esteja a mostra
   3- Clicar na guia 'Concentração de Negociação'
   4- Selecionar o campo de busca como 'CNPJ'

"""

import ctypes
import csv
import os

import pyautogui

import time
import datetime
from datetime import date
from datetime import timedelta
from datetime import datetime

# Carrega a lista de agentes filtradas do Infomercado
agentesfile=open('agentes.csv')
reader = csv.reader(agentesfile)
datalist=list(reader)
size=len(datalist[:][:]) #define a quantidade de agentes12.194.903/0001-30
pyautogui.hotkey('alt', 'tab')
time.sleep(0.5)

# Encontra a posição dos botões de agente e baixar
start = pyautogui.locateCenterOnScreen('start.png', confidence=0.9)
baixar = pyautogui.locateCenterOnScreen('baixar.png', confidence=0.9)

#Cria o log
log = open("Logs/resultados.log","w")
log.write(datetime.now().strftime("%Y-%m-%d %H:%M:%S")+" - INFO - Starting application\n")
log.write("*****************************************************************************************************************\n")
log.close()


log = open("Logs/resultados.log","a")
log.write(datetime.now().strftime("%Y-%m-%d %H:%M:%S")+" - INFO - Downloading files \n")
log.write("*****************************************************************************************************************\n")
log.close()

# Inicio do loop que vai baixar os dados caso tenha algo disponível para aquele agente
for j in range(21,2172):

    
    #Variável pra saber até aonde o código rodou
    print('j='+str(j))
    aux=[]
    
    #Pega o nome do agente pra renomear o arquivo e o CNPJ para filtrar no site
    agente=datalist[j][0]  
    CNPJ=datalist[j][1]
    
    
    #Escreve nome do agente e o horário no log criado
    log = open("Logs/resultados.log","a")
    log.write(datetime.now().strftime("%Y-%m-%d %H:%M:%S")+" - INFO - Baixando "+agente+"\n")
    log.close()
    
    #1) Insere o CNPJ do agente
    
    pyautogui.moveTo(start)   
    pyautogui.click()
    time.sleep(1)
    pyautogui.write(CNPJ)
    
    time.sleep(1)
    pyautogui.press('down')   
    time.sleep(1)
    pyautogui.press('enter')
    time.sleep(1.5)  
    
    #2) Checa se o dado existe ou não
    aux= pyautogui.locateCenterOnScreen('checkvazio.png', confidence=0.9)
    
    try:
        aux=len(aux) #tenta checar se o aux gravou um valor, caso contrário o except pula para a próxima iteração do for
    except:
        #Se não tiver dados, escreve no log que não teve, e continua para a próxima iteração
        log = open("Logs/resultados.log","a")
        log.write(datetime.now().strftime("%Y-%m-%d %H:%M:%S")+" - INFO - Sem dados\n")
        log.write("*****************************************************************************************************************\n")
        log.close()
        continue
    
    
    #3) Baixar os dados
    time.sleep(1)
    pyautogui.moveTo(baixar) #move para o botão baixar
    pyautogui.click()
    
    time.sleep(1)
    tabref = pyautogui.locateCenterOnScreen('tabref.png', confidence=0.9)
    pyautogui.moveTo(tabref) #move para o botão tabela de referência cruzada
    pyautogui.click()
    
    time.sleep(1)
    histconc = pyautogui.locateCenterOnScreen('histconc.png', confidence=0.8)
    pyautogui.moveTo(histconc) #move para o botão que seleciona os dados corretos 
    pyautogui.click()

    time.sleep(1)
    baixar_2 = pyautogui.locateCenterOnScreen('baixar_2.png')
    pyautogui.moveTo(baixar_2) #move para o botão de download 
    pyautogui.click()
       
    time.sleep(5)
    
    pyautogui.press('enter')
    
    time.sleep(3)
    
    #Renomeia o novo arquivo criado
    old_name = r"C:\Users\U342925\Downloads\Hist Concentração de Negociação (2).xlsx"
    new_name = f"C:\\Users\\U342925\\Downloads\\{agente}.xlsx"
    os.rename(old_name,new_name)
    
    
    time.sleep(1)
    
    #Fecha a janela de download para não atrapalhar o processo
    x = pyautogui.locateCenterOnScreen('x.png', confidence=0.9)
    pyautogui.moveTo(x) 
    pyautogui.click()
    
    #Escreve no log que os dados da empresa xxx foram baixados
    log = open("Logs/resultados.log","a")
    log.write(datetime.now().strftime("%Y-%m-%d %H:%M:%S")+" - INFO - Download OK\n")
    log.write("*****************************************************************************************************************\n")
    log.close()
    
    #Printa o nome do último agente verificado
    print(agente)

#Escreve no log que a aplicação finalizou
log = open("Logs/resultados.log","a")
log.write(datetime.now().strftime("%Y-%m-%d %H:%M:%S")+" - INFO - Ending application\n")
log.close()


#Abre um pop up informando que finalizou os dados
ctypes.windll.user32.MessageBoxW(0, "O download dos arquivos foram concluídos", "NC Energia",64)
```
